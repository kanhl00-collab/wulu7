### P0 Scanner Tests

- 11 unit test cases for the Scanner module-
- the first 5 test cases are used test the branch in procedure `number()`
- the rest 6 test cases are used to test the if the scanner get the correct symbol for float type numbers and double type numbers.

```python
import nbimporter; nbimporter.options["only_defs"] = False
from P0 import compileString
import SC
```

```python
from SC import TIMES, DIV, MOD, AND, PLUS, MINUS, OR, EQ, NE, LT, GT, \
    LE, GE, PERIOD, COMMA, COLON, NOT, LPAREN, RPAREN, LBRAK, RBRAK, \
    LARROW, RARROW, LBRACE, RBRACE, CARD, COMPLEMENT, UNION, INTERSECTION, \
    ELEMENT, SUBSET, SUPERSET, DOTDOT, THEN, DO, BECOMES, INTEGER, IDENT, \
    SEMICOLON, ELSE, IF, WHILE, CONST, TYPE, VAR, SET, PROCEDURE, PROGRAM, \
    INDENT, DEDENT, EOF, DOUBLE, FLOAT

symbol = {TIMES: 'TIMES', DIV: 'DIV', MOD: 'MOD', AND: 'AND', PLUS: 'PLUS',
    MINUS: 'MINUS', OR: 'OR', EQ: 'EQ', NE: 'NE', LT: 'LT', GT: 'GT',
    LE: 'LE', GE: 'GE', PERIOD: 'PERIOD', COMMA: 'COMMA', COLON: 'COLON',
    NOT: 'NOT', LPAREN: 'LPAREN', RPAREN: 'RPAREN', LBRAK: 'LBRAK', RBRAK: 'RBRAK',
    LARROW: 'LARROW', RARROW: 'RARROW', LBRACE: 'LBRACE', RBRACE: 'RBRACE',
    CARD: 'CARD', COMPLEMENT: 'COMPLEMENT', UNION: 'UNION', INTERSECTION: 'INTERSECTION',
    ELEMENT: 'ELEMENT', SUBSET: 'SUBSET', SUPERSET: 'SUPERSET', DOTDOT: 'DOTDOT',
    THEN: 'THEN', DO: 'DO', BECOMES: 'BECOMES', INTEGER: 'INTEGER', IDENT: 'INDENT',
    SEMICOLON: 'SEMICOLON', ELSE: 'ELSE', IF: 'IF', WHILE: 'WHILE', CONST: 'CONST',
    TYPE: 'TYPE', VAR: 'VAR', SET: 'SET', PROCEDURE: 'PROCEDURE', PROGRAM: 'PROGRAM',
    INDENT: 'INDENT', DEDENT: 'DEDENT', DOUBLE: 'DOUBLE', FLOAT: 'FLOAT'}

def scanString(src): # for a more readable scanner output
    SC.init(src); syms = []
    while SC.sym != SC.EOF:
        syms.append((symbol[SC.sym], SC.newline))
        SC.getSym()
    return syms
```

#### Recogonize number of double type

```python
scanString("""
var g: double
procedure q(v: double) 
  var l: double       
    l := 9.55
    if l > v then writeDouble(l)
    else writeDouble(g)
program p
  g := 5.89; q(7.99)
""")
```

#### Recogonize number of float type

```python
scanString("""
program
var y : float
y := 5.0f
if a then
  if b then
    a := b
x := 3.2f
""")
```

#### Recogonize numbers of float and double type

```python
scanString("""
program p
  while 2.78 > 3.77f do
    write(1.000001)
""")
```

#### Recogonize array of double type

```python
scanString("""
type T = [1..10] → double
var a: T
procedure r()
    a[3] := 9.88
program p
  a[3] := 9.99
""")
```
